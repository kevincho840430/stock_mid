# Load_data 

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
df_all = pd.concat([df_train, df_test])
# df_all.sort_values('date').reset_index(drop=True).reset_index(drop=True, inplace=True)
```

# Add new feature

```python
## weekday
# 生星期幾的特徵
df_all['weekday'] = df_all['date'] % 7 
```

```python
#將大家的漲跌和交易額加起來
df_all["sum_Turnover"] = 0
df_all["sum_ups_and_downs"] = 0

for c in df_all.columns:
    if "Turnover" in c:
        df_all["sum_Turnover"] += df_all[c]
    elif "ups_and_downs" in c:
        df_all["sum_ups_and_downs"] += df_all[c]
```

# Split data

```python
train_cols = df_all.drop(columns=['date','0056_close_y']).columns

#train split
X, y =  _df_train[train_cols], _df_train['0056_close_y']
```

```python
#X.head()
```

```python
#y
```

# Model

# 1.xgboost

```python
from xgboost.sklearn import XGBRegressor as xgb

_model = xgb(n_estimators=500)
_model.fit(X, y, 
            eval_metric = 'rmse',
            verbose=False)
```

```python
_test = _df_test.drop(columns=['date', '0056_close_y'])
preds = _model.predict(_test)
#put the predict result to the test
_df_test['pred'] = preds
```

```python
#preds
```

# check answer and predict

```python
from sklearn import metrics
sol = pd.read_csv("../Taichung 5-mid/new_data/solution.csv")
sol = sol.merge(_df_test[['date','pred']],  how='left', on ='date')

#calculate mse
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(sol['0056_close_y'], sol['pred'])))
# np.sqrt( metrics.mean_squared_error(sol['0056_close_y'], scalar.inverse_transform(sol['pred'].values.reshape(-1,1)) ))
```

```python
#train 裡面的最大值只有395
#我們的預測基本上卻不到那個程度
#但股票通常有機會漲的比較高
#所以我們需要linear的模型輔助
print(np.max(y))
sol['pred'].plot()
```

# Tuning the model

# 2.xgboost -> xgboost(gblinear)>75%

```python
# add booster,reg_alpha(L1),eta(learning rate)
_model = xgb(booster="gblinear", reg_alpha=1, eta=0.1, n_estimators=500)
# _model = xgb(booster="gblinear", eta=0.001, estimators=500)
_model.fit(X, y, 
            eval_metric = 'rmse',
            verbose=False)
# _model.fit(X.values[:, f_index], y, 
#             eval_metric = 'rmse',
#             verbose=False)
```

```python
y.describe()
```

```python
#將大於75百分位數之值都改用linear model預測
#這樣比較大的值，不會因為在同個結點而被迫平均
r_index = np.where(_model.predict(_test) > 351.675916)[0]
sol.iloc[r_index, -1] = _model.predict(_test)[r_index]

print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(sol['0056_close_y'], sol['pred'])))
```

# Add y value be a feature

```python
#將之前預測的值加入當作feature
#但training有看過答案
#所以我們直接拿實際y就可
#但想拿預測也行
X.loc[:, 'y'] = y.values
X.loc[:, 'y'] = X['y'].shift(1)
X.at[0, 'y']  = X.at[1, 'y']

#填值
_test['y'] = sol['pred'].values
#預測為下一天所以shift
_test['y'] = _test['y'].shift(1)
#nan補上一天
_test.iat[0, -1] = X.iat[-1, -1]
```

```python
_test
```

```python
#預測
_model = xgb(n_estimators=500)
_model.fit(X, y, 
            eval_metric = 'rmse',
            verbose=False)
```

```python
preds = _model.predict(_test)
_df_test.loc[:, 'pred'] = preds
```

```python
from sklearn import metrics
sol = pd.read_csv("../Taichung 5-mid/new_data/solution.csv")
sol = sol.merge(_df_test[['date','pred']],  how='left', on ='date')

print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(sol['0056_close_y'], sol['pred'])))
```

## 3.xgboost -> xgboost(gblinear)>75%(training) -> xgboost(gblinear)>75%(tree based)+ predict 'y' value

```python
_model = xgb(booster="gblinear", reg_alpha=1, eta=0.1, n_estimators=500)
_model.fit(X, y, 
            eval_metric = 'rmse',
            verbose=False)
```

```python
#可以看到這次tree based 預測結果很棒
#他已經能逼近比較高的值
#所以我們這次不取training的75%
#改取tree based 預測出來的75%
sol.pred.describe()
```

```python
r_index = np.where(_model.predict(_test) > 391.956360)[0]
sol.iloc[r_index, -1] = _model.predict(_test)[r_index]

print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(sol['0056_close_y'], sol['pred'])))
```
